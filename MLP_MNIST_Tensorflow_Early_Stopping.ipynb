```python
import tensorflow.compat.v1 as tf

tf.disable_v2_behavior()

# import tensorflow as tf -> placeholder 에러남 (2.0버전 이후)
```

```python
X = tf.placeholder(tf.float32, shape = [4,2]) # input : 4, value : 2개
y = tf.placeholder(tf.float32, shape = [4,1]) # 결과값 : 4개, 1
```

```python
# first layer-첫번째 hidden layer

W1 = tf.Variable(tf.random_uniform([2,2])) # x1, x2 2개의 값을 받고, 2개의 node
B1 = tf.Variable(tf.zeros([2])) # 2개의 node에 들어감
z = tf.sigmoid(tf.matmul(X,W1) + B1) # activation function 
```

```python
# second layer

W2 = tf.Variable(tf.random_uniform([2,1])) # 첫번째 hidden layer 값이 두번째 layer에 들어옴 , 1개의 결과값을 냄
B2 = tf.Variable(tf.zeros([1])) # 노드가 1개라서 바이어스도 1개
Y_hat = tf.sigmoid(tf.matmul(z, W2) + B2) # 예측값
```

```python
# Loss function - cross entropy (0 또는1 출력)
# cf) linear에서는 보통 MSE
# classfication에서는 cross entropy 사용함
loss = tf.reduce_mean(-1*((y*tf.log(Y_hat)) + ((1-y)*tf.log(1.0-Y_hat))))
```

```python
# optimizer - Gradient Descent

train_step = tf.train.GradientDescentOptimizer(0.05).minimize(loss)
    # loss function을 최소화하는 방향으로 optimizer 할 것
```

```python
# Train 
train_X = [[0,0], [0,1], [1,0], [1,1]]
train_Y = [[0], [1], [1], [0]]
```

```python
# initialize
    # 변수 초기화_tf에서 변수형은 그래프를 실행하기 전에 초기화 해야 값이 변수에 지정됨
init = tf.global_variables_initializer()
# start traning
with tf.Session() as sess:
    # run the initializer
    sess.run(init)
    print("train data: " + str(train_X))
    for i in range(20000):
        sess.run(train_step, feed_dict = {X : train_X, y : train_Y})
        if i % 5000 == 0:
            print("Epoch : " , i)
            print("Output : ", sess.run(Y_hat, feed_dict = {X : train_X, y : train_Y}))
        
        print("Final Output : ", sess.run(Y_hat, feed_dict = {X : train_X, y : train_Y}))
        
# 0,0  1,1 일 때 0에 가까움
# 1,0  0,1 일 때 1에 가까움
# 50% 이상일 땐 1을 output
# 50% 이하일 땐 0을 output
#  ==> XOR operation을 잘 실행
```

# MLP(MNIST)

```python
# MLP(MNIST)

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
print(x_train.shape) # 60000개, 28 * 28 pixels :: 1개의 row에 28개의 컬럼이 있음
print(x_test.shape)
```

```python
# train data를 train data와 validation data로 구분
# validation data - validation data로 나온 점수를 통해 hyperparameter 수정 가능 
#                 - overfitting을 막기위해 validation accuracy가 진전되지 않을 때 early stopping을 하여 
#                     training time 줄이고 최적의 모델 생성

x_val = x_train[50000:60000]
x_train = x_train[0:50000]
y_val = y_train[50000:60000]
y_trainl = y_train[0:50000]


#print("train data has " + str(x_train.shape[0]) + " sample")
#print("every train data is " + str(x_train.shape[1]) + " * " + str(x_train.shape[2]) + " image")
```

```python
x_train[0] # 하나의 array에 784개의 숫자가 있음
```

```python
# Normalize data - 정규화

x_train = x_train.astype('float32')
x_val = x_val.astype('float32')
x_test = x_test.astype('float32')

gray_scale = 255
x_train /= gray_scale
x_val /= gray_scale
x_test /= gray_scale
```

```python
# label to one hot encoding value

num_classes = 10 # 10개의 노드
# 0부터 9까지의 숫자를 10개의 노드로 넣기위해 one-hot encoding
y_train = tf.keras.utils.to_categorical(y_train, num_classes) # 모든 y값들을 one-hot-encoding으로 변경
y_val = tf.keras.utils.to_categorical(y_val, num_classes)
y_test = tf.keras.utils.to_categorical(y_test, num_classes)
```

```python
y_train
```

```python
# Tensorflow MLP graph

x = tf.placeholder(tf.float32, [None, 784]) # 몇 개의 sample을 쓸진 모르지만 총 784개를 input
y = tf.placeholder(tf.float32, [None, 10]) # 10개의 숫자를 구분 y:output layer
# placeholder : 진짜 value는 아니지만 그래프에 해당형태의 data가 들어갈 것이라고 말하는 것 
```

```python
def mlp(x):
    # hidden layer1
    w1 = tf.Variable(tf.random_uniform([784, 256])) # 총 784개의 input을 받고, 256개의 node가 있음
    b1 = tf.Variable(tf.zeros([256])) # 총 256개의 bias, 각 node마다 bias가 있음
    h1 = tf.nn.relu(tf.matmul(x, w1) + b1) # activation function
    
    # hidden layer2
    w2 = tf.Variable(tf.random_uniform([256, 128]))
    b2 = tf.Variable(tf.zeros([128]))
    h2 = tf.nn.relu(tf.matmul(h1, w2) + b2)
    
    # ouput layer
    w3 = tf.Variable(tf.random_uniform([128, 10]))
    b3 = tf.Variable(tf.zeros([10]))
    logits = tf.matmul(h2 , w3) + b3
    
    return logits
```

```python
logits = mlp(x)
```

```python
# cross entropy로 soft max 많이 사용
loss_op = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits= logits, labels = y))
```

```python
train_op = tf.train.AdamOptimizer(learning_rate= 0.01).minimize(loss_op)
# optimizer가 모든 variable을 최적화함
```

```python
# tensorflow 그래프를 직접 돌려봄
#initialize

init = tf.global_variables_initializer()

# train hyperparameters
epoch_cnt = 30
batch_size = 1000 # 1000개씩 넣으면서 아키텍쳐를 최적화
iteration = len(x_train) // batch_size

# Start training
with tf.Session() as sess:
    # Run the initializer
    sess.run(init)
    for epoch in range(epoch_cnt):
        avg_loss = 0
        start = 0; end = batch_size
        
        for i in range(iteration): # 미니배치, 
            # 1000개마다 돌아옴 1000개 데이터씩 iteration 돌아감
            _, loss = sess.run([train_op, loss_op],
                           feed_dict = {x : x_train[start:end], y : y_train[start:end]})
                            # 실제데이터가 들어감
            start += batch_size; end += batch_size
            # compute average loss
            avg_loss += loss / iteration
            
        # Validate model
        preds = tf.nn.softmax(logits) # apply softmax to logits
            # prediction value
        correct_prediction = tf.equal(tf.argmax(preds, 1), tf.argmax(y, 1))
        # Calculate accuracy
        accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
        cur_val_acc = accuracy.eval({x : x_val, y : y_val})
        print("epoch : " + str(epoch) + ", validation accuracy : " + str(cur_val_acc) + ", loss : "+ str(avg_loss))
    
    # Test model
    preds = tf.nn.softmax(logits) # apply softmax to logits
    correct_prediction = tf.equal(tf.argmax(preds, 1), tf.argmax(y,1))
    # Caculate accuracy
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
    print("Test accuracy : ", accuracy.eval({x : x_test, y: y_test}))
    
    # early stopping - overffiting을 막기위해, 훈련 시간을 낭비하지 않기위함
    # epoch을 많이 했다고해서 최적의 model이라고 할 수는 없음
    # --> training data에만 최적화 되어있을수도 있기때문
    # --> validation data의 accuracy가 더이상 증가하지 않을 때 
    # --> training data에 대해서는 점점 증가할 것 (계속 최적화되므로)
    # --> training accuracy는 증가하는데 validation accuracy는 증가하지 않을 때 overfitting된다고 볼 수 있음
    # ---> 이럴 때 early stopping 하고 최적의 모델선정
```

```python
# early stopping

# initailize
init = tf.global_variables_initializer()

# add ops to save and restore all the variables
saver = tf.train.Saver()

# trina hyperparameters
epoch_cnt = 300
batch_size = 1000
iteration = len(x_train) // batch_size

earlystop_threshold = 5 # 5로 선택 => valdidation accuracy가 올라갔다가 내려오는경우도 있음
        # validation score가 바로 떨어졌다고해서 학습을 stop하는 것은 있을 수 없는 일
        # 5번 기다리겠다고 이야기함 - 모델에 따라 바꾸기 (5번동안 꾸준히 안올라가면 스탑)
earlystop_cnt = 0
```

```python
# strat training 
with tf.Session() as sess:
    sess.run(init)
    prev_train_acc = 0.0
    max_val_acc = 0.0
    
    for epoch in range(epoch_cnt):
        avg_loss = 0.
        start = 0; end = batch_size
        
        for i in range(iteration):
            _, loss = sess.run([train_op, loss_op],
                           feed_dict = {x : x_train[start:end], y : y_train[start:end]})
            start += batch_size; end += batch_size
            # compute average loss
            avg_loss += loss / iteration
            
        # validation model
        preds = tf.nn.softmax(logits) # apply softmax to logits
        correct_prediction = tf.equal(tf.argmax(preds, 1), tf.argmax(y,1))
        # Calculate accuracy
        accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
        # train accuracy
        cur_train_acc = accuracy.eval({x: x_train, y: y-train})
        # validation accuracy
        cur_val_acc = accuracy.eval({x : x_val, y: y_val})
        # validation loss
        cur_val_loss = loss_op.eval({x : x_val, y: y_val})
        
        print("epoch : "+ str(epoch) + ", train acc : " + str(cur_train_acc)) 
        
        
        if cur_val_acc < max_val_acc:
            if cur_train_acc > prev_train_acc or cur_train_acc > 0.99:
                if earlystop_cnt == earlystop_threshold:
                    print("early stopped on " + str(epoch))
                    break
                else:
                    print("overfitting waring : " + str(earlystop_cnt))
                    earlystop_cnt += 1
            else:
                earlystop_cnt = 0
        else:
            earlystop_cnt = 0
            max_val_acc = cur_val_acc
            # save th variables to file
            save_path = saver.save(sess, "model/model.ckpt")
        prev_train_acc = cur_train_acc
    
```
